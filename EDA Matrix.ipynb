```python
# Heat Map
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
data = np.random.randint(low=1, high=100,size=(10,10))
print("The data to be plotted:\n")
print(data)
hm = sn.heatmap(data = data)
plt.show()
```

```python
# Heat Map
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
data = np.random.randint(low=1, high=100,size=(10,10))
annot = True
print("The data to be plotted:\n")
print(data)
hm = sn.heatmap(data = data, annot = annot)
plt.show()
```

```python
# Heat Map
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
data = np.random.randint(low=1, high=100,size=(10,10))
cmap = "tab10"

annot = True
print("The data to be plotted:\n")
print(data)
hm = sn.heatmap(data = data, cmap=cmap,annot=annot)
plt.show()
```

```python
linewidths = 2
linecolor = "yellow"

hm = sn.heatmap(data = data, cmap=cmap,annot=annot,linewidths=linewidths,linecolor=linecolor)
```
