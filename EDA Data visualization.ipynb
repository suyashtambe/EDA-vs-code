```python
# ax.bar
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,24,19,20,22]
ax.bar(langs,students);
```

```python

data = [30, 25, 50, 20], [40, 23, 51, 17], [35, 22, 45, 19]
X = np.arange(4)

print(X)
print(data)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, data[0], color = 'b', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(X + 0.50, data[2], color = 'r', width = 0.25);
```

```python
N = 5
menMeanscore = (20, 35, 30, 35, 27)
womenMeanscore = (25, 32, 34, 20, 25)
ind = np.arange(N)
width = 0.35
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(ind, menMeanscore, width, color='r')
ax.bar(ind, womenMeanscore, width, bottom=menMeanscore,color='b')
ax.set_ylabel('scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind, ('G1','G2','G3','G4','G5'))
ax.set_yticks(np.arange(0,81,10))
ax.legend(labels=['Men', 'Women'])
plt.show()
```

```python
fig,ax = plt.subplots(1,1)
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
ax.hist(a, bins = [0,25,50,75,100])
ax.set_title("histogram of result")
ax.set_xticks([0,25,50,75,100])
ax.set_xlabel('marks')
ax.set_ylabel('no.of students')
plt.show()
```

```python
#creating dataset
np.random,seed(356586)
N_points = 10000
n_bins = 20

# creating distribution
x = np.random.randn(N_points)
y = .8 ** x + np.random.randn(10000) + 25

# creating histogram
fig, axs = plt.subplots(1, 1, figsize = (10, 7), tight_layout = True)
axs.hist(x, bins = n_bins)

plt.show()
```
