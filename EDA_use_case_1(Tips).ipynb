```python
# Uni-variate Analysis
sns.histplot(tips['total_bill'],kde=False);
```

```python
# Bivariate Analysis
sns.jointplot(x="total_bill",y="tip",data=tips)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,hue="smoker")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,hue="sex")
```

```python
# which gender has given the most tip
sns.barplot(data=tips, x="sex", y="tip")
```

```python
male = 0
female = 0
for i in range(len(tips["sex"])):
    if tips["sex"][i] == "Male":
        male += tips["total_bill"][i]
    else:
        female += tips["total_bill"][i]
print("Total bill for Male:", male)
print("Total bill for Female:", female)
```

```python
#Multivariate Analysis
sns.pairplot(tips, hue='sex', markers=["o", "s"], diag_kind='hist')
plt.title("Pair Plot of Tips Dataset")
plt.show()

# Example: Correlation heatmap
correlation_matrix = tips.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
